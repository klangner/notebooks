# Probability distributions

Initialize imports

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Exponential distribution

```python
df = pd.Series(np.random.exponential(scale=1, size=500))

fig = plt.figure(figsize=(16,4))
ax1 = fig.add_subplot(121)
ax1.set_title("Exponential distribution with scale=1")
ax1.set_xlabel('random value')
ax1.set_ylabel('count')
df.hist(bins=100)
ax2 = fig.add_subplot(122)
ax2.set_title("Random values from exponential distribution")
ax2.set_xlabel('some random numbers')
ax2.set_ylabel('more random numbers')
df.plot()
```

## Poisson distribution with single lambda

```python
l = 15
df2 = pd.Series(np.random.poisson(lam=l, size=500))

fig = plt.figure(figsize=(16,4))
ax1 = fig.add_subplot(121)
ax1.set_title("Poisson distribution with lambda=%d" % l)
ax1.set_xlabel('random value')
ax1.set_ylabel('count')
df2.hist(bins=100)
ax2 = fig.add_subplot(122)
ax2.set_title("Random values from poisson distribution")
ax2.set_xlabel('some random numbers')
ax2.set_ylabel('more random numbers')
df2.plot()
```

## Poisson distribution with 2 different lambdas

```python
l1 = 15
l2 = 10
df3 = pd.Series(np.append(np.random.poisson(lam=l1, size=100), np.random.poisson(lam=l2, size=10)))

fig = plt.figure(figsize=(16,4))
ax1 = fig.add_subplot(121)
ax1.set_title("Poisson distribution with 2 lambdas")
ax1.set_xlabel('random value')
ax1.set_ylabel('count')
df3.hist(bins=100)
ax2 = fig.add_subplot(122)
ax2.set_title("Random values from poisson distribution")
ax2.set_xlabel('some random numbers')
ax2.set_ylabel('more random numbers')
df3.plot()
```

## Binomial distribution

```python
df4 = pd.Series(np.random.binomial(n=1, p=0.5, size=500))

fig = plt.figure(figsize=(16,4))
ax1 = fig.add_subplot(121)
ax1.set_title("Binomial distribution with 1 trial")
ax1.set_xlabel('random value')
ax1.set_ylabel('count')
df4.hist(bins=100)
ax2 = fig.add_subplot(122)
ax2.set_title("Random values from binomial distribution")
ax2.set_xlabel('some random numbers')
ax2.set_ylabel('more random numbers')
df4.plot()
```
