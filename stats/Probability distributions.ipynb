# Probability distributions

Initialize imports

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def plot_dist(df, name):
    fig = plt.figure(figsize=(12,4))
    ax1 = fig.add_subplot(121)
    ax1.set_title("%s distribution" % name)
    ax1.set_xlabel('random value')
    ax1.set_ylabel('count')
    plt.hist(df, bins=100)
    ax2 = fig.add_subplot(122)
    ax2.set_title("Random values from %s distribution" % name)
    ax2.set_ylabel('Random number')
    plt.plot(df)
```

## Exponential distribution

```python
df = np.random.exponential(scale=5, size=500)
plot_dist(df, "Exponential")
```

## Poisson distribution with single lambda

```python
df2 = np.random.poisson(lam=5, size=500)
plot_dist(df2, "Poisson")
```

## Poisson distribution with 2 different lambdas

```python
l1 = 15
l2 = 10
df3 = np.append(np.random.poisson(lam=l1, size=100), np.random.poisson(lam=l2, size=10))
plot_dist(df3, "Poisson")
```

## Binomial distribution

```python
df4 = np.random.binomial(n=1, p=0.5, size=500)
plot_dist(df4, "Binomial")
```

## Uniform distribution

```python
df5 = np.random.uniform(size=500)
plot_dist(df5, "Uniform")
```
